### JSON Files

```python
# Create a dictionary for each book with info such as author, genre, etc.
book1 = {'title': 'The Prophet',
         'author': 'Khalil Gibran',
         'genre': 'poetry',
         'tags': ['religion', 'spirituality', 'philosophy', 'Lebanon', 'Arabic', 'Middle East'],
         'book_id': '811.19',
         'copies': [{'edition_year': 1996,
                     'checkouts': 486,
                     'borrowed': False},
                    {'edition_year': 1996,
                     'checkouts': 443,
                     'borrowed': False}]
         }
         
book2 = {'title': 'The Little Prince',
         'author': 'Antoine de Saint-Exupery',
         'genre': 'children',
         'tags': ['fantasy', 'France', 'philosophy', 'illustrated', 'fable'],
         'id': '843.912',
         'copies': [{'edition_year': 1983,
                     'checkouts': 634,
                     'borrowed': True,
                     'due_date': '2017/02/02'},
                    {'edition_year': 2015,
                     'checkouts': 41,
                     'borrowed': False}]
         }

library = [book1, book2]
library
```

```python
import json
# Write the library of books to the file 'library.json'
with open('./library.json', 'w') as f:
    json.dump(library, f, indent=2)
```

```python
# Read the json file using the json library
with open('./library.json', 'r') as f:
    reloaded_library = json.load(f)

reloaded_library
```

### Beautiful Soup to Scrape Wikipedia Site

```python
# Use BeautifulSoup library to extract the list of works by Vincent Van Gogh for example
from bs4 import BeautifulSoup
import requests
url = requests.get("http://en.wikipedia.org/wiki/List_of_works_by_Vincent_van_Gogh").text
soup = BeautifulSoup(url, 'lxml')
table = soup.find('table', class_="wikitable sortable")
# table
links = table.findAll('a')
# links

# List of works
works = []
for link in links:
    works.append(link.get('title'))
# Remove the None values
Van_Gogh_Works = []
for work in works:
    if work != None:
        Van_Gogh_Works.append(work)

# Unique values using the set function
Van_Gogh_Works = list(set(Van_Gogh_Works))
# Create a data frame
Van_Gogh_Works_df = pd.DataFrame({'Work': Van_Gogh_Works})
# len(Van_Gogh_Works)

# Extract the first few rows of the dataframe
Van_Gogh_Works_df.head()
```
